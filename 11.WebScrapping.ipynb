# C. Create a review scrapper for any ecommerce website to fetch real time comments, reviews, ratings, comment tags, customer name using Python.

### Product: Samsung Galaxy S25 5G — Flipkart

- **Product:** Samsung Galaxy S25 5G (Icyblue, 256 GB) | 12 GB RAM
- **Link:** https://www.flipkart.com/samsung-galaxy-s25-5g-icyblue-256-gb/p/itm3c1e495903acc

## Step 1: Import Required Libraries

```python
import requests
import bs4
import pandas as pd
```

## Step 2: Set Headers and URL

```python
# Mimic a real browser to avoid getting blocked
headers = {"User-Agent": "Mozilla/5.0"}
```

```python
# Samsung Galaxy S25 5G product URL on Flipkart
url = "https://www.flipkart.com/samsung-galaxy-s25-5g-icyblue-256-gb/p/itm3c1e495903acc"
```

## Step 3: Send HTTP Request & Parse HTML

```python
response = requests.get(url, headers=headers)
print("Status Code:", response.status_code)
```

```python
soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

## Step 4: Scrape Product Title

```python
# Product title is inside an <h1> tag with Flipkart's class
title = soup.find('h1', {'class': 'v1zwn21l v1zwn26 _1psv1zeb9 _1psv1ze0'})
print("Product Name :", title.get_text(strip=True))
```

## Step 5: Scrape Product Price

```python
# Price is inside a <div> tag
price = soup.find('div', {'class': 'v1zwn21l v1zwn20 _1psv1zeb9 _1psv1ze0'})
print("Price :", price.get_text(strip=True))
```

## Step 6: Scrape Overall Product Rating

```python
# Overall rating shown prominently on the page
ratings = soup.find_all('div', {'class': 'css-146c3p1'})
print("Overall Product Rating :", ratings[1].text)
```

## Step 7: Scrape Individual Review Ratings

```python
# Each review has a star rating badge
individual_ratings = soup.find_all('div', class_='_1psv1zeb9 _1psv1ze0 _1psv1zee9 _1psv1zel0')

for rating in individual_ratings:
    print("Rating:",rating.text[0],"⭐")
```

## Step 8: Scrape Review Comment Tags (Short Headlines)

```python
# Comment tags are the short one-liners summarising each review
comment_tags = soup.find_all('div', class_= '_1psv1zeb9 _1psv1ze0 _1psv1zee9 _1psv1zel0')

for comment in comment_tags:
    print("Comment:",comment.text[1:])
```

## Step 9: Scrape Customer Names

```python
filtered_customers = []

customer_name_tags = soup.find_all('div', class_= 'v1zwn21l v1zwn27 _1psv1zeb9 _1psv1ze0')

# Filter out non-customer-name strings like delivery info, services, etc.
unwanted_prefixes = [
    "10 Days", "Cash on", "Delivery", "Flipkart", "Assured",
    "Right Product", "No delivery", "Quick service", "by",
    "Free test", "Nearby", "All accessories", "Expert assistance",
    "Guaranteed help", "No hassle", "Quality checks","brand support",
    "6 step guidance", "Authorised service", "Expert help","128 GB + 12 GB"
]

for tag in customer_name_tags:
    text = tag.get_text(strip=True)

    if text == "Flipkart Customer" or not any(
        word in text for word in unwanted_prefixes
    ):
        filtered_customers.append(text)

for name in filtered_customers:
    print("Customer:", name)
```

## Step 10: Count Total Review Blocks

```python
# vQDoqR is the class for each review card block
review_blocks = soup.find_all('div', class_='vQDoqR')
print("Total Review Blocks Found:", len(review_blocks))
```

## Step 11: Scrape Full Review Text

```python
# Full review body text for each review
reviews = soup.find_all('div', class_='v1zwn21l v1zwn26 _1psv1zeb9 _1psv1ze0')

for review in reviews:
    print("Review:",review.get_text(strip=True))
```

## Step 12: Combine Data into a DataFrame & Export to CSV

```python
data = []

for customer, rating, comment, review in zip(
    filtered_customers,
    individual_ratings,
    comment_tags,
    reviews
):

    data.append({
        "Customer Name": customer,
        "Rating": rating.text[0],
        "Comment Tag": comment.text[1:],
        "Review": review.get_text(strip=True)
    })

df = pd.DataFrame(data)

df.to_csv("samsung_s25_flipkart_reviews.csv", index=False)

print("Data saved successfully")
```
