# C. Create a review scrapper for any ecommerce website to fetch real time comments, reviews, ratings, comment tags, customer name using Python.

- link: https://www.flipkart.com/motorola-80-cm-32-inch-qled-hd-ready-smart-google-tv/p/itm247d6554b407b

```python
import requests
import bs4
```

```python
headers = {"User-Agent": "Mozilla/5.0"}
```

```python
url = "https://www.flipkart.com/motorola-80-cm-32-inch-qled-hd-ready-smart-google-tv/p/itm247d6554b407b"
```

```python
request1 = requests.get(url, headers=headers)
```

```python
soup = bs4.BeautifulSoup(request1.text, 'html.parser')
```

```python
#Title
title = soup.find('h1', {'class':'v1zwn21l v1zwn26 _1psv1zeb9 _1psv1ze0'})

if title:
    print("Product Name :", title.get_text(strip=True))
```

```python
#Price
price = soup.find('div',{'class':'v1zwn21l v1zwn20 _1psv1zeb9 _1psv1ze0'})
if price:
    print("Price :", price.get_text(strip=True))
```

```python
#Overall Rating
rating = soup.find_all('div', {'class':'css-146c3p1'})

print("Overall Product Rating :", rating[1].text)
```

```python
# Rating
rating_tags = soup.find_all('div',class_='_1psv1zeb9 _1psv1ze0 _1psv1zee9 _1psv1zel0')

for rating in rating_tags:
    print("Rating:",rating.text[0])
```

```python
# comments
comment_tags = soup.find_all('div',{'class':'_1psv1zeb9 _1psv1ze0 _1psv1zee9 _1psv1zel0'})

for comment in comment_tags:
    print("Comment:",comment.text[1:])
```

```python
filtered_customers = []
customer_name = soup.find_all('div',{'class':'v1zwn21l v1zwn27 _1psv1zeb9 _1psv1ze0'})

for cust_name in customer_name:
    text = cust_name.get_text(strip=True)
    unwanted = ["10 Days","Cash on","Delivery","Flipkart","Assured","Right Product","No delivery","Quick service","by","Free test","Nearby","All accessories","Expert assistance","Guaranteed help","No hassle","Quality checks","6 step guidance","Authorised service","Expert help"]

    if text == "Flipkart Customer":
        filtered_customers.append(text)

    elif (
        not any(text.startswith(word) for word in unwanted)
        and len(text.split()) <= 3
    ):
        filtered_customers.append(text)

for customer in filtered_customers:
    print(customer)
```

```python
# Get all review blocks
reviews = soup.find_all('div', class_='vQDoqR')
print("Total Reviews:", len(reviews))
```

```python
reviews = soup.find_all('div',class_='v1zwn21l v1zwn26 _1psv1zeb9 _1psv1ze0')

for review in reviews:
    print(review.get_text(strip=True))
    print()
```

```python
import pandas as pd
import re

data = []

ratings = soup.find_all('div',class_='_1psv1zeb9 _1psv1ze0 _1psv1zee9 _1psv1zel0')

reviews = soup.find_all('div',class_='v1zwn21l v1zwn26 _1psv1zeb9 _1psv1ze0')

# -------- STORE DATA --------
for i in range(min(len(filtered_customers), len(ratings), len(reviews))):

    customer = filtered_customers[i]
    rating_text = ratings[i].get_text(strip=True)
    rating = re.search(r'\d+(\.\d+)?', rating_text).group()
    review = reviews[i].get_text(strip=True)

    data.append({
        "Customer Name": customer,
        "Rating": rating,
        "Review": review
    })

df = pd.DataFrame(data)

df.to_csv("flipkart_reviews.csv", index=False)

print("Web scraping completed!")
```
