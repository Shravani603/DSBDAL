# 8. Visualize the data using Python libraries matplotlib, seaborn by plotting the graphs for airquality dataset.(Use Air_quality_forvisualization.csv)( Charts : Line chart, Barplot, Heatmap, Scatterplot, histogram, boxplot, violin, timeseries chart)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AQ_data.csv")   # or your file name
df.head()
```

```python
df.shape
```

```python
df.columns
```

# Line Chart

```python
plt.figure()
plt.plot(df.index[:30], df['so2'][:30], color='steelblue')
plt.title("SO2 Levels (First 30 Records)")
plt.xlabel("Index")
plt.ylabel("SO2")
plt.show()
```

# Bar Plot

```python
#Average SO₂ by State
plt.figure()
sns.barplot(data=df, x='state', y='so2')
plt.xticks(rotation=90)
plt.title("Average SO2 by State")
plt.xlabel("State")
plt.ylabel("SO2")
plt.show()
```

# Heatmap

```python
# Correlation between numerical features.
plt.figure(figsize=(8,5))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

# Scatter Plot

```python
# Relationship between NO2 vs PM2.5 (colored by type)
plt.figure()
sns.scatterplot(data=df, x='no2', y='pm2_5', hue='type')
plt.title("NO2 vs PM2.5")
plt.xlabel("NO2")
plt.ylabel("PM2.5")
plt.legend(title="type", bbox_to_anchor=(1,1), loc='upper left')
plt.show()
```

# Histogram

```python
# Distribution of PM2.5
plt.figure(figsize=(8,6))
sns.histplot(df['pm2_5'], bins=30, kde=True, color="green")
plt.title("Histogram: PM2.5 Distribution")
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="type", y="no2", data=df)
plt.xticks(rotation=90)
plt.title("Box Plot: NO₂ Levels by Area Type")
plt.xlabel("Area Type")
plt.ylabel("NO₂ Concentration")
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(10,6))
sns.violinplot(x="type", y="no2", data=df, inner="box", palette="muted")
plt.xticks(rotation=90)
plt.title("Violin Plot: NO₂ Distribution by Area Type")
plt.xlabel("Area Type")
plt.ylabel("NO₂ Concentration")
plt.show()
```

# Time Series Chart

```python
df_small = df.iloc[:100]

plt.figure()
plt.plot(df_small.index, df_small['no2'], color='steelblue')
plt.title("NO2 Time Series (First 100 Records)")
plt.xlabel("Index")
plt.ylabel("NO2")
plt.show()
```
