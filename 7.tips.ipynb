# 7. Visualize the data using Python libraries matplotlib, seaborn by plotting the graphs for tips dataset.( Charts : Line chart, Barplot, Heatmap, Scatterplot, histogram, boxplot, violin, timeseries chart)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("tips (2).csv")   # or your file name
df
```

```python
df = df.drop_duplicates()
```

# Line Chart

```python
# Tip increases as no of size increases
plt.figure()
sns.lineplot(data=df, x='size', y='tip')
plt.title("Size vs Tip (Line Chart)")
plt.xlabel("Size")
plt.ylabel("Tip")
plt.show()
```

- The line chart shows that tips generally increase as the size increases

# Bar Plot

```python
# Which day has the highest average tips.
plt.figure()
sns.barplot(data=df, x='day', y='total_bill', palette='Set1')
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

- The bar plot shows that Total Bill vary across days, with higher tips observed on weekends

# Heatmap

```python
# Correlation between numerical features.
plt.figure(figsize=(8,5))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

- The heatmap shows a strong positive correlation between total bill and tipm

# Scatter Plot

```python
# Relationship between total bill and tip (colored by smoker).
# Smoker's differences in tipping behavior.
plt.figure()
sns.scatterplot(data=df, x='total_bill', y='tip', hue='smoker')
plt.title("Total Bill vs Tip (Smoker vs Non-Smoker)")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

- The scatter plot shows a positive relationship between total bill and tip, with slight variation between smokers and non-smokers.

# Histogram

```python
# Distribution of total bill amounts.
plt.figure()
sns.histplot(data=df, x='total_bill', bins=20, kde=True, color="green")
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Count")
plt.show()
```

- The histogram shows that most total bills fall within a moderate range, with fewer very high values

# Box Plot

```python
# Tip distribution across different days.
plt.figure()
sns.boxplot(data=df, x='day', y='tip', palette='Set2')
plt.title("Tips Distribution by Day")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

- The box plot shows variation in tips across days and highlights the presence of outliers.

# Violin Plot

```python
# total bill distribution across different time (dinner, lunch).
plt.figure()
sns.violinplot(data=df, x='time', y='total_bill', palette='Set3')
plt.title("Total Bill Distribution by Time")
plt.xlabel("Time")
plt.ylabel("Total Bill")
plt.show()
```

- The violin plot shows that dinner bills tend to be higher and more varied compared to lunch.

# Time Series Chart

```python
# how total_bill varies overtime across the first 50 observations
df_small = df.iloc[:50]

plt.figure()
plt.plot(df_small.index, df_small['total_bill'], color="purple")
plt.title("Time Series of Total Bill (First 50 Records)")
plt.xlabel("Index")
plt.ylabel("Total Bill")
plt.show()
```

- The time series plot shows fluctuations in total bill amounts across observations without a clear trend.

```python
# Trend of tips across dataset entries.
plt.figure(figsize=(8,6))
plt.plot(df.index, df['tip'])
plt.title("Time Series Chart: Tips Across Records")
plt.xlabel("Index (Order of Records)")
plt.ylabel("Tip")
plt.show()
```
